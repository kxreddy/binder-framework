```python
df=pd.read_csv("C:\\DataFiles\\Value_of_Energy_Cost_Savings_Program_Savings_for_Businesses_20231225.csv")
```

```python
cols=[df.columns.str.replace(' ','_')]
cols
df.columns=['Period', 'Company_Name', 'company_contact', 'company_email',
        'company_phone', 'Address', 'City', 'State', 'Postcode', 'Industry',
        'Industry_descr', 'Company_Type', 'Current_fulltime', 'Job_created',
        'Job_retain', 'Effective_Date', 'Total_Savings',
        'Savings_from_beginning_receiving_benefits', 'Gas_Savings',
        'Cogen_savings', 'Electric_Savings', 'Borough', 'Latitude', 'Longitude',
        'Community_Board', 'Council_District', 'BIN', 'BBL',
        'Census_Tract_(2020)', 'Neighborhood_Tabulation_Area_(NTA)_(2020)']
df
```

```python
#How many different companies are represented in the data set?
len(df.groupby('Company_Name'))
```

```python
#What is the total number of jobs created for businesses in Queens?
b=df.groupby(['Borough'])['Job_created'].sum()
b.xs('QUEENS')
```

```python
#How many different unique email domains names are there in the data set?
len(df['company_email'].str.split('@').str[1].unique())
```

```python
#aggregate NTS by count of companies
df_revised=df.groupby(['Neighborhood_Tabulation_Area_(NTA)_(2020)']).agg({'Company_Name':'count'})
df_revised
```

```python
#filater list of NTA with > 5 companies 
filt_5=df_revised.index[df_revised['Company_Name']>5]
filt_5
```

```python
dfidx_rev=df.set_index('Neighborhood_Tabulation_Area_(NTA)_(2020)') #Set index as NTA to search the dataframe based NTA
df_5CompNTA=dfidx_rev.loc[filt_5] #use filter to retrive only NTAs with > 5 companies
df_5CompNTA=df_5CompNTA.reset_index() #reset index
df_5CompNTA
```

```python
#aggregate the total jobs created and average total saving for NTA with >5 companies
df_agg_tot=df_5CompNTA.groupby(['Neighborhood_Tabulation_Area_(NTA)_(2020)']).agg({'Company_Name':'count','Job_created':'sum','Total_Savings':'mean'})
df_agg_tot.rename(columns={'Company_Name':'count_of_Companies','Job_created':'Total_jobs_created','Total_Savings':'Total_average_saving'},inplace=True) #change the column name as meaningful
df_agg_tot
```

```python
#Save your result for the previous question as a CSV file
df_agg_tot.to_csv('C:\\DataFiles\\DIpandas.csv')

```
