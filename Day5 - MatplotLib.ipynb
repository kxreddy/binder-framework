```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mlb
import numpy as np
```

```python
df_nta=pd.read_csv('C:\\DataFiles\\DIpandas.csv')
df_nta
```

```python
x=df_nta['Total_jobs_created'].tolist()
y=df_nta['Total_average_saving'].tolist()
log_y=np.log(df_nta['Total_average_saving'])
log_y=log_y.tolist()
```

```python
#Scatter plot with Standard average savings values
N = len(x)

c = np.random.randint(1, 5, size=N)
s = np.random.randint(10, 220, size=N)

fig, ax = plt.subplots(figsize=(6, 3), layout='constrained')
#ax.scatter('a', 'b', c='c', s='d', data=data)
#ax.scatter(x,log_y,c=c,s=s)
ax.scatter(x,y,c=c,s=s)
ax.set_xlabel('Total Jobs Created')
ax.set_ylabel('Total Avg Savings')
ax.set_title("Jobs Created Vs Avg Savings")  # Add a title to the axes.
#ax.legend('Total Jobs Created','Total Avg Savings')  # Add a legend.
```
